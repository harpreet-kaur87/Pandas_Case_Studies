## Find all employees with a Performance Rating of 5 and Salary less than 50,000.

```python
performance_analysis = df[(df['PerformanceRating'] == 5) & (df['Salary'] < 50000)]
performance_analysis
```

## Find employees who joined the company after January 1, 2015.

```python
employee_joined_after_2015 = df[df['JoinDate'] > '2015-01-01']
employee_joined_after_2015
```

## What is the average salary of employees in each department?

```python
avg_sal = df.groupby('Department')['Salary'].mean()
avg_sal
```

## What is the total sales amount for each region?

```python
total_sales_per_region = df.groupby('SalesRegion')['SalesAmount'].sum()
total_sales_per_region
```

## Find the number of employees in each department.

```python
df.Department.value_counts()
```

## What is the maximum salary in each department?

```python
max_salary_per_dept = df.groupby('Department')['Salary'].max()
max_salary_per_dept
```

## Find the total bonus paid in each department.

```python
total_bonus_per_dept = df.groupby('Department')['Bonus'].sum()
total_bonus_per_dept
```

## Sort the dataset by Salary in descending order.

```python
sorted_data = df.sort_values(by='Salary', ascending=False)
sorted_data
```

## Sort the employees by Performance Rating in descending order and Name in alphabetical order.

```python
new_order_rating = df.sort_values(by=['PerformanceRating','Name'], ascending=[False,True])
new_order_rating
```

## Sort the employees by Age, and then by Salary within each age group.

```python
grouping_by_age = df.sort_values(by=['Age','Salary'], ascending=[True,False])
grouping_by_age
```

## Find the second-highest salary in each department.

```python
second_highest_salary = (
                df.groupby('Department')['Salary']
               .apply(lambda x:x.nlargest(2).iloc[-1] if len(x) > 1 else None).reset_index()
                .rename(columns={'Salary':'Second_Highest_Salary'})
)
second_highest_salary
```

## Convert the 'JoinDate' column to a datetime format if it is not already.

```python
df['JoinDate'] = pd.to_datetime(df['JoinDate'])
df.info()
```

## Extract the year from the 'JoinDate' and create a new column 'JoinYear'.

```python
df['JoinYear'] = df['JoinDate'].dt.year
df.head(5)
```

## Create a new column 'AgeGroup' where employees above 50 are 'Senior' and others are 'Junior'.

```python
df['AgeGroup'] = df['Age'].apply(lambda x:'Senior' if x > 50 else 'Junior')
df.head(5)
```

## Calculate the percentage of each departmentâ€™s total salary in the overall salary expenditure.

```python
department_salary = df.groupby('Department')['Salary'].sum()
total_salary = department_salary.sum()
```

```python
department_salary_percentage = (department_salary / total_salary) * 100
department_salary_percentage
```

## Find the employees who are in the 'Sales' department, have a salary greater than 80,000, and joined after 2018.

```python
data_from_sales_dept = df[(df['Department'] == 'Sales') & (df['Salary'] > 80000) & (df['JoinYear'] > 2018)]
data_from_sales_dept
```

## Find all employees with a performance rating of 4 or 5 who have received a bonus greater than 7,000.

```python
high_performance_rating = df[(df['PerformanceRating'] >= 4) & (df['Bonus'] > 7000)]
high_performance_rating
```

## Find the average sales amount by region for employees with a performance rating greater than 3.

```python
avg_sales_amount = df[df['PerformanceRating'] > 3].groupby('SalesRegion')['SalesAmount'].mean()
avg_sales_amount
```

## What is the average salary and total sales amount by department?

```python
dept_analysis = df.groupby('Department')[['Salary','SalesAmount']].agg({'Salary':'mean','SalesAmount':'sum'})
dept_analysis
```

## Find the department with the highest total salary expenditure.

```python
dept_with_highest_sal_exp = df.groupby('Department')['Salary'].sum().idxmax()
dept_with_highest_sal_exp
```

## Find the total salary expenditure and the number of employees in each department where the total salary exceeds 2 million.

```python
department_summary = df.groupby('Department').agg({'Salary':'sum','EmployeeID':'count'})
department_summary[department_summary['Salary'] > 2000000]
```

## Calculate the median salary by department and performance rating.

```python
median_salary = df.groupby(['Department','PerformanceRating'])['Salary'].median()
median_salary
```

## Find the department-wise distribution of employee performance ratings (frequency of each rating).

```python
dist_of_rating = df.groupby(['Department','PerformanceRating']).size().unstack(fill_value=0)
dist_of_rating
```

## Rank employees within each department based on their sales amount. If two employees have the same sales amount, rank them by salary.

```python
df['SalesRank'] = ( 
            df.sort_values(['SalesAmount','Salary'], ascending=[False,False]).groupby('Department')
            .cumcount() + 1
)
```

```python
df.sort_values(by=['Department','SalesRank'], ascending=[True,True])
```
